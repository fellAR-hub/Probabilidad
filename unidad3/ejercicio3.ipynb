```python
import numpy as np
import matplotlib.pyplot as plt


valores = np.array([0, 1, 2, 3, 4, 5])
probabilidades = np.array([0.05, 0.10, 0.25, 0.30, 0.20, 0.10])

suma_prob = np.sum(probabilidades)
print("Suma de probabilidades:", suma_prob)


print("\nDistribución de probabilidad:")
for x, p in zip(valores, probabilidades):
    print(f"X = {x}, P(X) = {p}")


E = np.sum(valores * probabilidades)
print("\nValor esperado E(X):", E)

varianza = np.sum(probabilidades * (valores - E)**2)
print("Varianza:", varianza)

desv_std = np.sqrt(varianza)
print("Desviación estándar:", desv_std)


F = np.cumsum(probabilidades)

print("\nFunción de distribución acumulada:")
for x, f in zip(valores, F):
    print(f"F({x}) = {f}")


plt.step(valores, F, where='post')
plt.xlabel("Unidades vendidas")
plt.ylabel("F(x)")
plt.title("Función de Distribución Acumulada")
plt.grid()
plt.show()
```
