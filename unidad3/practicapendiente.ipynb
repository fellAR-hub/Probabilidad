```python
import numpy as np
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5]
p=[0.10,0.25,0.30,0.20,0.10,0.05]
print("suma de pobabilidades",sum(p))

plt.stem(x,p)
plt.xlabel("numero de celulas anormales encontradas")
plt.ylabel("probabilidad")
plt.grid(True)
plt.show()
```
