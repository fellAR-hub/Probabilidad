```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70  # Media
desv_std = 4  # Desviación estándar
total_estudiantes = 500

# VARIABLE para cálculo de la probabilidad de que un estudiante pese menos de 75 kg
peso_limite = 75

# Probabilidad usando la CDF
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)

# Número de estudiantes
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

# Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)

plt.title('Probabilidad de que un estudiante pese menos de 75 kg.\n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

# Parámetros de la distribución normal
media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 62

prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Número de estudiantes que pesan menos de 62 kg:", estudiantes_menos_62kg)

# Graficar
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)

plt.title('Probabilidad de que un estudiante pese menos de 62 kg.\n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


media = 70
desv_std = 4
peso_limite = 79
total_estudiantes = 500


prob_menos_79kg = norm.cdf(peso_limite, loc=media, scale=desv_std)


prob_mas_79kg = 1 - prob_menos_79kg


estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("Probabilidad de que un estudiante pese más de 79 kg:", round(prob_mas_79kg, 6))
print("Número aproximado de estudiantes que pesan más de 79 kg:", round(estudiantes_mas_79kg, 1))

plt.figure(figsize=(8, 5))


x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, loc=media, scale=desv_std)


plt.plot(x, y, 'b', linewidth=2)


plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)


plt.title('Probabilidad de que un estudiante pese más de 79 Kg.\n'
          f'$\mu={media}$, $\sigma={desv_std}$, $n={total_estudiantes}$',
          fontsize=14)
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)


plt.show()
```
