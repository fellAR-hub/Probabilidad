###ejercicio 3

```python
import numpy as np
import matplotlib.pyplot as plt
n=10
p=0.10
vx=np.arange(0,n+1)
px=[binom.pmf(x,n,p)for x in vx]
print("valores de x\tprobabilidades")
for x,probabilidad in zip(vx,px):
  print(f"{x}\t\t{probabilidad}")
```

```python
plt.figure(figsize=(8,6))
plt.bar(vx,px,color='skyblue')
plt.title('distribucion binomial')
plt.xlabel('valor de x')
plt.ylabel('probabilidad')
plt.xticks(vx)
plt.grid(True)
plt.show
```
