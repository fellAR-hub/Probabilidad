```python
import random

import pandas as pd
import statistics as stats
edades=[]
for i in range(100):
  na=random.randint(18,70)
  edades.append(na)
print(edades)
print("================================================")
df=pd.DataFrame(edades,columns=["edades"])
df.head()
#---------------------------------------------
tabfre=df['edades'].value_counts().reset_index()
tabfre.columns=["edades","frecuencia"]
tabfre=tabfre.sort_values(by='edades')
print("tabla de distribucion de frecuencias")
tabfre
print("================================================")
tabfre["frecuencia relativa"]=tabfre['frecuencia']/100
tabfre
#--------------------
tabfre["frecuencia porcentual"]=tabfre["frecuencia"]/100*100
tabfre
#-----------------
tabfre["frecuencia acumulada"]=tabfre["frecuencia"].cumsum()
tabfre
```
