```python
import random
import math
import pandas as pd

edades=[]
for i in range(100):
  NA=random.randint(18,70)
  edades.append(NA)
print(edades)
df=pd.DataFrame(edades,columns=['edad'])
df.head()
tf=df['edad'].value_counts().reset_index()
tf.columns=['edad','frecuencia']
tf=tf.sort_values(by='edad')
print('tabla de distribucion de frecuencias:')
tf['Frec Relativa']=tf['frecuencia']/100
tf['Frec porcentual']=tf['frecuencia']/100*100
tf['Frec acumulada']=tf['frecuencia'].cumsum()
tf


```

```python
df=pd.DataFrame({'Edad':edades})
n=len(edades)
k=int(1 + math.log2(n))


print('valor de n:',n,'\nNumero de intervalos=',k)

df
```

```python
import matplotlib.pyplot as plt

tf=pd.cut(df['Edad'],bins=k,include_lowest=True).value_counts().reset_index()
tf.columns=['clase','Frec']
tf=tf.sort_values(by='clase')
tf['Frec Rel'] = tf['Frec'] / n
tf['Frec Porc'] = tf['Frec'] / n * 100
tf['Frec Ac'] = tf['Frec'].cumsum()
tf['Frec Rel Ac'] = tf['Frec Rel'].cumsum()
tf['Frec Porc Ac'] = tf['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia
print(tf)
plt.hist(edades,alpha=0.5)
plt.title("histogramas de edades basados en la funcion de matplotlib con valores estandar")
```
