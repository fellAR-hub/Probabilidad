```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV en una variable llamada covid
covid = pd.read_csv("20-06-30-COVID19MEXICO - codificado.csv", encoding="ISO-8859-1")

# Mostrar los primeros 5 registros
print(covid.head())

# Mostrar información general del archivo
print(covid.info())

# Edad máxima y mínima
print("Edad máxima:", covid["EDAD"].max())
print("Edad mínima:", covid["EDAD"].min())

# ========================
# HISTOGRAMA DE EDAD
# ========================

media = covid["EDAD"].mean()

plt.figure()
plt.hist(covid["EDAD"], bins=11)
plt.axvline(media, linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')
plt.title("Histograma Edad Pruebas COVID")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

# ========================
# HISTOGRAMA TIPO PACIENTE
# ========================

plt.figure()
plt.hist(covid["TIPO_PACIENTE"], bins=3)
plt.title("Histograma Tipo Paciente en Pruebas COVID")
plt.xlabel("Tipo de Paciente")
plt.ylabel("Frecuencia")
plt.show()

# ========================
# HISTOGRAMA SEXO
# ========================

plt.figure()
plt.hist(covid["SEXO"], bins=3)
plt.title("Histograma Sexo en Pruebas COVID")
plt.xlabel("Sexo")
plt.ylabel("Frecuencia")
plt.show()

# ========================
# BOXPLOT POR SEXO
# ========================

plt.figure()
sns.boxplot(x="SEXO", y="EDAD", data=covid)
plt.title("Distribución de Edad por Sexo")
plt.show()

# ========================
# BOXPLOT POR TIPO DE PACIENTE
# ========================

plt.figure()
sns.boxplot(x="TIPO_PACIENTE", y="EDAD", data=covid)
plt.title("Distribución de Edad por Tipo de Paciente")
plt.show()

# ========================
# BOXPLOT POR RESULTADO
# ========================

plt.figure()
sns.boxplot(x="RESULTADO", y="EDAD", data=covid)
plt.title("Distribución de Edad por Resultado")
plt.show()
```
