```python
import random

import  matplotlib.pyplot as plt
from statistics import *

fabrica1=[]
for i in range (100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
print("=====================")
fabrica2=[]
for i in range (100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
print("=====================")
maximo1=max(fabrica1)
minimo1=min(fabrica1)
rango1=maximo1-minimo1
media1=mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)
def1=stdev(fabrica1)
print("-----fabrica 1-----")
print("maximo1:",maximo1)
print("minimo1:",minimo1)
print("rango1:",rango1)
print("media:",media1)
print("mediana1:",mediana1)
print("moda1:",moda1)
print("desviacion estandar 1:",def1)
print("=====================")
print("-----fabrica 2-----")
maximo2=max(fabrica2)
minimo2=min(fabrica2)
rango2=maximo2-minimo2
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)
def2=stdev(fabrica2)
print("maximo2:",maximo2)
print("minimo2:",minimo2)
print("rango2:",rango2)
print("media2:",media2)
print("mediana2:",mediana2)
print("moda2:",moda2)
print("desviacion etandar:",def2)
```

```python
plt.figure(figsize=(8,6))
plt.hist(fabrica1,bins=18,alpha=0.2,color="b",edgecolor="black")
plt.axvline(media1,color="b",linestyle="dashed",linewidth=2,label=f'media={media1:.2f}')
plt.axvline(mediana1,color="r",linestyle="dashed",linewidth=2,label=f'mediana={mediana1:.2f}')
plt.axvline(moda1,color="y",linestyle="dashed",linewidth=2,label=f'moda={moda1:.2f}')
plt.axvline(media1-def1,color="black",linestyle="dashed",linewidth=2,label=f'desv.estandar{def1:.2f}')
plt.axvline(media1+def1,color="black",linestyle='dashed',linewidth=2)
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.title('produccion diaria de fabrica 1')
plt.legend()
plt.show()
import seaborn as sns
sns.histplot(x=fabrica1,kde=True,kde_kws={'bw_adjust':0.5},bins=18,color="y")
```
