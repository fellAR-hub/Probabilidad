```python
from statistics import *
grupoa=[20,87,98,99,100,100,50,68,70,80,84,24]
print(grupoa)
type(grupoa)
promedio=sum(grupoa)/len(grupoa)
print(promedio)
promedio2=mean(grupoa)
print(promedio2)
mediana=median(grupoa)
print(mediana)
moda=mode(grupoa)
print(moda)
moda2=multimode(grupoa)
print(moda2)
```

```python
tiempos100mts=[14.05,22.52,66.12,43.10,11.11,15.67,30.88,25.34,12.44,53.41]
print(tiempos100mts)



print("tiempo promedio de los corredores:",mean(tiempos100mts))
print("tiempo mediano de los corredores:",median(tiempos100mts))
print("tiempo de la moda de los corredores:",mode(tiempos100mts))
```

```python
import random
fabrica1=[]
for i in range(100):
  numero_aleatorio = round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
print("[-------------------------------]")
fabrica2=[]
for i in range(100):
  numero_aleatorio = round(random.uniform(50,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
plt.hist(fabrica1)
plt.title("produccion diaria de la fabrica 1")
print("================================")
```

```python
plt.hist(fabrica2)
plt.title("produccion diaria de la fabrica 2")
```

```python
print("valores de fabrica1:")
print("la media es:",mean(fabrica1))
print("la mediana es:",median(fabrica1))
print("la moda es:",mode(fabrica1))
print("============================")
print("valores de fabrica 2:")
print("la media es:",mean(fabrica2))
print("la mediana es:",median(fabrica2))
print("la moda es:",mode(fabrica2))
```
